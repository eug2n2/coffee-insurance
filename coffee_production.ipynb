```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('1a - Total production.csv')
df
```

```python
df.columns #열 확인
```

```python
df.drop(df.index[0],inplace = True)
df
```

```python
df.dropna(inplace=True)
```

```python
for col in df.columns[2:32]:
    try:
        df[col] = df[col].str.replace(',', '').astype('int64')
    except ValueError as e:
        print(f"에러 발생: {e} (열 {col})")
```

```python
df
```

```python
df.drop('Unnamed: 1',axis= 1,inplace = True)
```

```python
# 새로운 열 이름 리스트 생성
new_columns = ['Total production by all exporting countries']
for year in range(1990, 2020):
    new_columns.append(f"{year}/{year + 1}")

# 열 이름을 새로운 이름으로 변경
df.columns = new_columns

df
```

```python
df.to_csv('each_country_production.csv')
```

```python
df_melted = df.melt(id_vars=['Total production by all exporting countries'], var_name='Year', value_name='Total')

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Total', data=df_melted, hue='Total production by all exporting countries')
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Total')
plt.title('Total Production by All Exporting Countries Over the Years')
plt.show()
```

```python
# 각 열의 합을 계산하여 새로운 데이터프레임 생성
df_sum = pd.DataFrame({'Year': df.columns[1:], 'Total': df.sum(axis=0)[1:]})

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Total', data=df_sum)
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Total')
plt.title('Total Production by Year (Sum of All Countries)')
plt.show()
```

```python
# 국가별 그래프 생성
for i, row in df.iterrows():
    country_name = row['Total production by all exporting countries']
    country_data = row.drop('Total production by all exporting countries')

    # Total 값에 따라 색상을 생성
    color_palette = sns.color_palette("Blues", n_colors=len(country_data))
    colors = [color_palette[i] for i in range(len(country_data))]

    # 그래프 그리기
    plt.figure(figsize=(12, 6))
    sns.barplot(x=country_data.index, y=country_data.values, palette=colors)
    plt.xticks(rotation=90)
    plt.xlabel('Year')
    plt.ylabel('Total')
    plt.title(f'Total Production for {country_name} by Year (Shades of Blue)')
    plt.show()
```

```python
top_10_countries = df.sort_values(by='2019/2020', ascending=False).head(10)

# 결과 출력
top_10_countries
```

```python
# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='Total production by all exporting countries', y='2019/2020', data=top_10_countries, palette="Blues")
plt.xlabel('Total for 2019/2020')
plt.ylabel('Country')
plt.title('Top 10 Countries by Total Production for 2019/2020')
plt.show()
```

```python
# 파이 차트 그리기
plt.figure(figsize=(8, 8))
plt.pie(top_10_countries['2019/2020'], labels=top_10_countries['Total production by all exporting countries'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # 원형을 유지하도록 설정
plt.title('Top 10 Countries by Total Production for 2019/2020')
plt.show()
```

```python
pastel_colors = sns.color_palette("pastel", len(top_10_countries))

# 그래프 설정
plt.figure(figsize=(8, 8))
plt.pie(top_10_countries['2019/2020'],
        labels=top_10_countries['Total production by all exporting countries'],
        autopct='%1.1f%%', startangle=140, colors=pastel_colors)
plt.axis('equal')
plt.title('Top 10 Countries by Total Production for 2019/2020')

# seaborn 스타일 적용
sns.set(style="whitegrid")

# 그래프 보여주기
plt.show()
```
